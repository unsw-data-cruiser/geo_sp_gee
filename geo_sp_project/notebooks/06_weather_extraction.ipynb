# 06 · Historical Weather Extraction (Open-Meteo)

Extracts 2-year daily weather for all Beachwatch sites with checkpoint/resume.

> **Prerequisite:** run `05_beachwatch_registry` first to create `bw_site_registry_basic.csv`

```python
import sys
if '/content/geo_sp_project/src' not in sys.path:
    sys.path.extend(['/content/geo_sp_project/src', '/content/geo_sp_project/configs'])
```

```python
import pandas as pd, config
from geo_sp.open_meteo_weather import run_extraction

df_sites = pd.read_csv(config.SITE_REGISTRY_CSV)
print(f'Sites loaded: {len(df_sites)}')

cfg = dict(
    start_date            = config.WEATHER_START_DATE,
    end_date              = config.WEATHER_END_DATE,
    api_base_url          = 'https://archive-api.open-meteo.com/v1/archive',
    timezone              = config.WEATHER_TIMEZONE,
    request_delay         = config.WEATHER_REQUEST_DELAY,
    rate_limit_wait       = config.WEATHER_RATE_LIMIT_WAIT,
    max_retries           = config.WEATHER_MAX_RETRIES,
    checkpoint_interval   = config.WEATHER_CHECKPOINT_INTERVAL,
    checkpoint_file       = config.WEATHER_CHECKPOINT_FILE,
    temp_data_file        = config.WEATHER_TEMP_FILE,
    final_data_file       = config.WEATHER_FINAL_FILE,
)
```

```python
# ⚠️  This takes ~20-60 min for all sites.
# If interrupted, re-run this cell — it will resume from checkpoint.
df_weather = run_extraction(df_sites, cfg)

if not df_weather.empty:
    print(f'\n✅ {len(df_weather):,} records  |  {df_weather["site_id"].nunique()} sites')
    print(df_weather[['site_id','site_name','date','temp_2m_C','rain_24h_mm']].head(3))
```
