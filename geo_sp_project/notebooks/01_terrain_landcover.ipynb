```python
import config
from geo_sp.boundaries import get_lga, get_suburb, get_lga_info
from geo_sp.terrain import load_terrain_layers, elevation_stats, slope_stats, slope_class_pcts, build_terrain_map

randwick = get_lga(config.LGA_NAME, config.LGA_LAYER)
info     = get_lga_info(randwick)
print(f"Area: {info['area_km2']:.2f} km²  |  Centre: {info['centroid']}")
```

```python
# Terrain statistics
layers  = load_terrain_layers()
e_stats = elevation_stats(layers['elevation'], randwick)
s_stats = slope_stats(layers['slope'], randwick)
classes = slope_class_pcts(layers['slope'], randwick)

print('Elevation (m):',
      f"min={e_stats.get('elevation_min',0):.1f}",
      f"max={e_stats.get('elevation_max',0):.1f}",
      f"mean={e_stats.get('elevation_mean',0):.1f}")
print('Slope (deg):',
      f"mean={s_stats.get('slope_mean',0):.1f}",
      f"max={s_stats.get('slope_max',0):.1f}")
for k, v in classes.items():
    print(f'  {k}: {v:.1f}%')
```

```python
# Interactive terrain map
m = build_terrain_map(layers, randwick, config.MAP_CENTER, config.MAP_ZOOM)
m
```

## Land Cover

```python
from geo_sp.landcover import load_worldcover, landcover_stats, build_landcover_map
from geo_sp.boundaries import get_suburb

suburbs = {}
for name in config.SUBURB_LIST:
    g = get_suburb(name, config.SUBURB_LAYER)
    if g: suburbs[name] = g
print(f'Loaded {len(suburbs)} suburbs')

wc    = load_worldcover()
stats = landcover_stats(wc, randwick)
print(f"Built-up: {stats['built_km2']:.2f} km² ({stats['built_pct']:.1f}%)")
print(f"Water:    {stats['water_km2']:.2f} km² ({stats['water_pct']:.1f}%)")
print(f"Trees:    {stats['tree_km2']:.2f} km² ({stats['tree_pct']:.1f}%)")
```

```python
m2 = build_landcover_map(wc, randwick, suburbs, config.MAP_CENTER, config.MAP_ZOOM)
m2
```
